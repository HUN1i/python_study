```python

drama=['13333','23333','33333','43333']
recall = []
for i in drama:
    
    if i.startswith('1'): # 1로 시작하는가
        recall.append(i) # recall에 집어넣기
    print(recall)

# list comprehension

drama=['1','2','3','4']
recall = [i for i in drama if i.startswith('3')]
print(f'{recall}')
# 앞에 변수값 지정 가능(i,i+1,i*2 등등)
```
