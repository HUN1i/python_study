```python
# true,false 넣어주는게 또 번거로울 수 있다.
# 이 때 기본값 사용
# def 함수명(전달값=기본값) 양식으로 사용
def fruit(is_vip=False):# 과일가게에 일반손님이 더 많을 것
    if is_vip==True:
        return 10000
    else:
        return 15000
    # 이렇게 두게 되면
fruit1=fruit(True)
fruit2=fruit()
fruit3=fruit()# 비워놓으면 자동으로 False가 채워짐
print(f"{fruit1},{fruit2},{fruit3}")
```
