```python
# 손님 관리를 위해 customer 수를 지정하려하는데 이게 너무 불규칙적임
# 가변인자(개수가 바뀔 수 있는 인자 사용)
# 전달값 앞에 별 찍기
# 별 찍으면 튜플 형태로 받는다
def visit(today,*customers):
    print(today)
    for customer in customers:
        print(customer)

visit(20220102,'나')
visit(20220202,'너')
print(f'{visit}')
# 가변인자는 함수 내에서 한 번만 사용 가능함
```
