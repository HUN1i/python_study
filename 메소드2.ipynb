```python
# startswith:글자가 어떠한 글자로 시작하는지 확인. true or false. 끝나는 건 endswith
# strip:괄호 안에 제거하려는 문장 넣으면 제거됨.안넣으면 공백 제거
# replace: 특정 글자 바꾸기
# find: 문자열 위치 확인
# center: n글자로 감싸기
# python 공식 페이지에 메소드 있음
letter='asd'
print(letter.startswith('a'))
print(letter.strip('as'))
print(letter.replace('as','asdd'))
```
